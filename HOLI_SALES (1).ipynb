# EXPLORATORY DATA ANALYSIS

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
```

```python
# plotting a bar chart for Gender and it's count
order = ['M', 'F']
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.countplot(x = 'Gender',data = df,order=order,palette=['lightblue', 'lightpink'])

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,palette=['lightgreen', 'lightyellow'])
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Age Group', y='Amount',data=sales_age,palette=['darkgreen', 'yellow','red','pink','grey','brown','purple'])
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender',palette=['lightgreen', 'lightyellow'])
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```
